```python
import pandas as pd
import numpy as np

CoD = pd.read_csv("Multiple Cause of Death, 1999-2014.csv")
```

```python
#Task One
CoD = CoD.rename(columns={"Crude Rate": "Crude_Rate", "Crude Rate Lower 95% Confidence Interval":"Crude_Rate_Lower_95%_Confidence_Interval", "Crude Rate Upper 95% Confidence Interval": "Crude_Rate_Upper_95%_Confidence_Interval", }, errors="raise")
```

```python
#Task Two - (At first, the only two columns recognized as float are Year and Population)
#CoD.info()
CoD = CoD.replace({"Suppressed": 0, "Unreliable" : 0})
#Converting accorting to {https://www.geeksforgeeks.org/change-data-type-for-one-or-more-columns-in-pandas-dataframe/?ref=lbp}
convert_dict = {'Deaths': int,
                'Crude_Rate': float, 
                "Crude_Rate_Lower_95%_Confidence_Interval": float, 
                "Crude_Rate_Upper_95%_Confidence_Interval": float
                }
CoD = CoD.astype(convert_dict)
#CoD.info()
```

```python
#Task 3:  number of deaths > 2000 deaths only show State and Year
thoudeat = CoD.query('Deaths > 2000')
print("States and Years where Deaths exceed 2000:")
print(thoudeat[['State', 'Year']])
```

```python
#Task 4: total number of deaths per each state across all years
Tot = CoD.groupby("State").sum()
Tot = Tot.rename(columns={"Deaths": "Total"}, errors="raise")
print(Tot[["Total"]])
```

```python
#Task 5:  Find all rows where the name of the state begin with the letter 'A'
AoD = CoD
i = 64
while (i < 816):
    AoD = AoD.drop(index=[i])
    i = i+1
AoD
```

```python
#Task 6: top five states with the largest population in any given year only show State and Population
Mos = CoD[["State", "Population"]]
Mos = Mos.groupby("State").max()
Mos = Mos.sort_values(by=['Population'], ascending=False)
print(Mos.head(5))
```

```python
#Task 7:  Find five years with the highest numbers of the deaths for the state with the highest population. In your results, show only "Year" and "Deaths"
#State = California
CaliD = CoD[CoD.State == "California"]
CaliD = CaliD[["Year", "Deaths"]]
CaliD = CaliD.sort_values(by=['Deaths'], ascending=False)
print(CaliD.head(5))
```
